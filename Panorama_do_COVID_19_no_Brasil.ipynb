# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados foram tirados repósitorio no Github do [OurWorldInData](https://ourworldindata.org/) que é uma organização sem fins lucrativos que coleta e fornece gratuitamente dados sobre principalmente "pobreza, doença, fome, mudança climática, guerra, riscos existenciais e desigualdade". Os dados mundiais sobre Covid-19 são atualizados diariamente pela organização que é uma das mais respeitadas no mundo relacionada ao fornecimento de dados. Você pode acessar esses dados públicos atualizados [aqui](https://github.com/owid/covid-19-data/tree/master/public/data).

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

# importar o dataset COVID-19 para um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

## Análise Inicial dos Dados

Antes de começarmos a explorar nosso dataset, é preciso saber com o que estamos lidando. Para começarmos a entender melhor esse aspecto, vamos verificar as dimensões do dataset, como os nossos dados estão estruturados.


```
#verificando as dimensões do Dataframe
print('Número de Linhas: ', df.shape[0])
print('Números de Colunas: ', df.shape[1])

#verificando as 5 primeiras entradas
df.head()
```

```
#extrair nome das colunas
df.columns
```

Outro detalhe importante que precisamos verificar e garantir que estão da forma que precisamos é o tipo das variáveis. Para isso, utilizaremos o método `dtypes`.

```
df.dtypes
```

Vamos verificar o percentual de valores ausentes em todo o nosso DataSet.

```
percentual = ((df.isnull().sum()/df.shape[0])*100).sort_values(ascending=False)
percentual = pd.DataFrame(percentual)
percentual.columns = ["Porcentagem (%)"]
percentual.columns.name = "Colunas"
percentual
```

Como vimos ao analisar o tipo das variáveis, verificamos que a coluna `date` não está em formato datetime. Para que possamos explorar essa variável ao máximo, vamos transformá-la para o formato datetime.

```
df.date = pd.to_datetime(df.date)
```

Após alterar o formato, vamos verificar qual a data mais recente do nosso dataset. Com essa informação, vamos verificar o país líder em casos naquele momento.

```
df.date.max()
```

```
#verificando total de casos
df.loc[df.date == '2020-07-23'].sort_values(by='total_cases',ascending=False)
```

Pudemos verificar que os países líderes em casos na data de 23 de Julho de 2020 eram:

* Estados Unidos - 3.970.906 casos
* Brasil - 2.227.514 casos
* Índia - 1.238.635 casos

Agora, vamos verificar os casos de morte.

```
df.loc[df.date == '2020-07-23'].sort_values(by='total_deaths',ascending=False)
```

Em relação às mortes, os países que lideravam na data de 22 de Julho de 2020 eram:

* Estados Unidos - 143.190
* Brasil - 82.771
* Grã-Bretanha - 45.501

Para melhorar a visualização, vamos plotar um gráfico de barras.

```
top5_total_mortes = df.loc[df.date == '2020-07-23', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(14,6))

sns.barplot(top5_total_mortes.location, top5_total_mortes.total_deaths, palette='Blues_d', ax=ax,)

ax.set_title('TOP 5 Países com mais mortes por COVID-19')
ax.set_xlabel('Países')
ax.set_ylabel('Total de Mortes')

plt.tight_layout()
```

### Como o vírus está progredindo?

Uma das maiores preocupações das organizações de saúde é a progressão do número de casos e mortes no mundo.

Uma das melhores formas de visualizarmos a evolução ao longo do tempo, é o gráfico de linhas. Por isso, vamos plotar um gráfico de linha para demonstrar e evolução de mortes no mundo.

```
plt.figure(figsize=(8,6))
sns.lineplot(data=df.loc[df.location == "World", "total_deaths"].reset_index(drop=True), linewidth=3, color='red', marker='P', markersize=15, markeredgecolor='w', markeredgewidth=0.5, markevery=[-1])
plt.tight_layout()
```

### Análise Exploratória para o Brasil

Agora que já temos uma noção dos acontecimentos no mundo, vamos focar nos acontecimentos no Brasil.


```
# Criar a copia de um dataframe apenas com os registros do Brasil
df_brasil = df.loc[df.location =='Brazil'].copy()
df_brasil.head()
```

```
print('Data do Primeiro Registro: ',df_brasil.date.min())
print('Data do Último Registro: ',df_brasil.date.max())
```

Acima, verificamos que nosso dataset inclui datas desde **31-12-2019** até **23-07-2020**.

Vamos verificar quando foram identificados o primeiro caso e a primeira morte no Brasil.

```
#primeira morte
df_primeira_morte = df_brasil[['location','date','new_cases','total_cases','new_deaths','total_deaths']]
df_primeira_morte.loc[df_primeira_morte.new_deaths != 0.0].head(1)
```

```
print('A primeira morte foi registrada em {}'.format(df_brasil.date.loc[4201]))
```

```
#primeiro caso
df_primeiro_caso = df_brasil[['location','date','new_cases','total_cases','new_deaths','total_deaths']]
df_primeiro_caso.loc[df_primeiro_caso.new_cases != 0.0].head(1)
```

```
print('O primeiro caso foi registrado em {}'.format(df_brasil.date.loc[4180]))
```

Uma informação interessante para investigarmos é quanto tempo levou desde o primeiro caso até a primeira morte.

```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte
from datetime import date
dias = df_brasil.date.loc[4201] - df_brasil.date.loc[4180]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(dias.days))
```

```
#o dia em que foi registrado maior numero de novas mortes por covid-19 no Brasil
df_brasil.loc[df_brasil.new_deaths == df_brasil.new_deaths.max()]
```

```
print('No dia {} foi registrado o maior numero de novas mortes por COVID-19.'.format(df_brasil.date.loc[4280]))
print('Nesta data foram registrados {} novas Mortes. '.format(df_brasil.new_deaths.loc[4280]))
```

Vamos plotar um gráfico que nos mostre **Taxa de Mortalidade no Brasil**.


```
df_brasil_data = df_brasil.copy()

df_brasil_data.set_index('date', inplace=True)
df_tx_mortalidade_br = (df_brasil_data.total_deaths / df_brasil_data.total_cases) * 100

fig, ax = plt.subplots(figsize=(18,6))
ax.plot(df_tx_mortalidade_br)
ax.set_title('Taxa de mortalidade no Brasil')
ax.set_ylabel('% de mortalidade')

plt.tight_layout()
plt.show()
```

Por fim, vamos plotar gráficos de linha para compararmos a evolução dos casos e mortes no Brasil.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(206))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```
